# Exploratory Data Analysis on Olympics Data.

## Olympics
#### The Summer Olympics games are a major international multi-sport-event normally held once every four years.The games were first held in 1896 in Athense,Greece,and were most recently the 2020 Summer Olympics held in 2021 in Tokyo,Japan.The International Olympics Committee(IOC) organises the games and oversees the host city's preparation.The WInter Olympics created after the success of the summer olympics.It is regarded as the greatest and largest multi-sport international events in the world.The 2024 Summer Olympics will be held in Paris,France for a third time.


## About the dataset
#### In this dataset  we have two files one is 'athletes_data' and another is 'region'.This dataset is avaliable till 2016 Rio olympics.

## Columns
#### 1. ID : Unique Id assigned to all available datapoints.
### # 2. Name : Name of the athletes.
#### 3. Sex : Gender of athletes.
#### 4. Age :Age of athlests at time of game.
#### 5. Height : Hieghts of participating athletes.
#### 6. Weight : Weights of participating athletes.
#### 7. Team : From which teams athletes are playing for.
#### 8. NOC : Country code name from which athletes are playing.
#### 9. Games : Summer/Winter with Year of Games.
#### 10.Year : Year of olympics game.
#### 11.Season : Season name of olympics(Summer/Winter)
#### 12. City: City of host olympics.
#### 13. Sport: Name of sport 
#### 14.Event : Events of Game
#### 15.Medal: If any athletes have won meda(Bronze,Silver,Gold)
#### 16.Region: The place from where athlest belong

```python
# Top Countries which r participating most

top_10_country=athletes_df.Team.value_counts().head(10)
top_10_country.head(10)
```

## Visualization

```python
plt.figure(figsize=(10,6))
plt.title('Top 10 Countries WRT to participants')
sns.barplot(x=top_10_country.index,y=top_10_country,palette='Set2');
```

```python
# Age Distribution of Participants

plt.figure(figsize=(12,6))
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('No of Athlets')
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color='Orange',edgecolor='white');
```

```python
winter_sports=athletes_df[athletes_df.Season=='Winter'].Sport.unique()
```

```python
summer_sports=athletes_df[athletes_df.Season=='Summer'].Sport.unique()
```

```python
summer_sports
```

```python
# Male and Female Participants

gender_counts=athletes_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True);
```

```python
# Total Medal
athletes_df.Medal.value_counts()
```

```python
female_parti_yearly=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_parti_yearly=female_parti_yearly.groupby('Year').count().reset_index()
female_parti_yearly
```

```python
Women_olympics=athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')]
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=Women_olympics,palette="Spectral")
plt.title("Women PArticipation");
```

```python
# Gold Medal

gold_medal=athletes_df[athletes_df.Medal=='Gold']
```

```python
gold_medal
```

```python
gold_medal['ID'][gold_medal['Age']>60].count()
```

```python
# GOld medalist country
gold_medal.region.value_counts()
```

```python
gold_medal.region.value_counts().reset_index(name='Medal').head(15)
```

```python
totalgoldmedal=gold_medal.region.value_counts().reset_index(name='Medal').head(8)
g=sns.catplot(x='index',y='Medal',data=totalgoldmedal,height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 8 countries')
g.set_ylabels(' Number of medal')
plt.title('Gold Medal Per Country')
```

```python
# Rio olympics
max_year=athletes_df.Year.max()
```

```python
max_year
```

```python
team_names=athletes_df[(athletes_df.Year==max_year)&(athletes_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20),y=team_names.value_counts().head(20).index)
plt.ylabel(None);
plt.xlabel('CountryWise Medals for the year 2016');
```

```python
not_null_medals=athletes_df[(athletes_df['Height'].notnull())&(athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize=(12,10))
axis=sns.scatterplot(x='Height',y='Weight',data=not_null_medals,hue='Sex')
plt.title('Height VS Weight of Olympics Medalists');
```
